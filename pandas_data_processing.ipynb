```python
import pandas as pd
```

```python
df=pd.read_csv("ashi_pyscho_data.csv")
df.head(10)
```

```python
df.isna().sum()
```

```python
df_new=df.dropna() #removing all NA values
```

```python
df_new.isna().sum()
```

```python
pd.set_option('display.max_columns', 10)
pd.set_option('display.max_rows', None)
#pandas display settings for clarity
```

```python
df_new=df_new.drop(df_new.columns[0], axis=1)
#dropping timestamp column cus not needed
df_new=df_new.reset_index(drop=True)
#resetting indexes
```

```python
df_new
```

```python
df_new.columns = df_new.columns.str.replace(' ', '_')
df_new.columns = df_new.columns.str.replace('(', '')
df_new.columns = df_new.columns.str.replace(')', '')
df_new.columns = df_new.columns.str.replace("’", "")
df_new.columns = df_new.columns.str.replace(',', '')
df_new.columns = df_new.columns.str.replace('.', '')
df_new
#removing space,commas and parenthesis for better parsing in python code and to avoid syntax errors
```

```python
mapping = {'First Year': 1, 'Second Year': 2, 'Third Year': 3, 'Fourth Year': 4}
df_new=df_new.replace({'I_am_currently_pursuing_my_degree_in_BTech_Degree_in_the_year:': mapping})
#changing catergorical data type to numerical for easier calculcatio
```

```python
df_new['Please_enter_your_current_CGPA_'] = df_new['Please_enter_your_current_CGPA_'].str.replace(r'[^0-9.]+', '')
df_new['Please_enter_your_percentage_percentage_only_from_grade_10th'] = df_new['Please_enter_your_percentage_percentage_only_from_grade_10th'].str.replace(r'[^0-9.]+', '')
df_new['Please_enter_your_percentage_percentage_only_from_grade_12th'] = df_new['Please_enter_your_percentage_percentage_only_from_grade_12th'].str.replace(r'[^0-9.]+', '')
#remove extra data which was entered incorrectly by form users using regex. only '.' and numbers are kept
```

```python
#Checking the types of data of columns
df_new.dtypes
```

```python
df_new.loc[30,:] #but we can see there's still some leftover data, this can be avoided by using regex implementation at form
```

```python
df_new.loc[40,:] #same case here, we will have to remove this manually
```

```python
df_new=df_new.drop(index=[30,40]) #dropping 30th and 40th row due to bad data
```

```python
df_new = df_new.apply(pd.to_numeric) #changing the dataframe type to numeric for calculations and pre-processing
```

```python
df_new=df_new.reset_index(drop=True) #resetting indexes again
df_new
```

```python
df_new=df_new.loc[(df_new.Please_enter_your_current_CGPA_ > 0) & (df_new.Please_enter_your_current_CGPA_ < 10)].reset_index(drop=True)
df_new=df_new.loc[(df_new.Please_enter_your_percentage_percentage_only_from_grade_10th > 10) & (df_new.Please_enter_your_percentage_percentage_only_from_grade_10th < 100)].reset_index(drop=True)
df_new=df_new.loc[(df_new.Please_enter_your_percentage_percentage_only_from_grade_10th > 10) & (df_new.Please_enter_your_percentage_percentage_only_from_grade_10th < 100)].reset_index(drop=True)
#fixing the cgpa and percentage columns, and remove data which is entered incorrently
```
